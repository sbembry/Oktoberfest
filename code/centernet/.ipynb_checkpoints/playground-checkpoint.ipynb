```python
import cv2
import numpy as np
import albumentations as A
import torch
from torch.utils.data import Dataset
import matplotlib.pyplot as plt


class OktoberfestDataset(Dataset):
    max_boxes = 10
    #show = False
    def __init__(self, lines, path, resize=(540, 960), augment=False, inference=False):
        """
        Creates Dataset with bboxes and labels
        lines: list of lines in csv containing img path and bboxes
        path: path to images
        resize: Resize image. Currently it is cutting image in half
        augment: boolean on whether to do augmentations or not
        inference: boolean on whether this is for inference or not. True returns normalized image
                   as model input and resized image, False returns processed image + bboxes.
        """
        self.lines = lines
        self.path = path
        self.transform = A.Compose([A.HorizontalFlip(p=.5),
                                    A.Rotate(limit = 30, 
                                             border_mode = cv2.BORDER_CONSTANT, 
                                             value = 0.0, p = 0)
                                    ], bbox_params=A.BboxParams(format='pascal_voc'))
        if not inference:
            self.resize = A.Compose([A.Resize(resize[0], resize[1], interpolation=cv2.INTER_AREA)],
                                     bbox_params=A.BboxParams(format='pascal_voc'))
        else:
            self.resize = A.Resize(resize[0], resize[1], interpolation=cv2.INTER_AREA)
        self.augment = augment
        self.normalize = A.Normalize()
        self.inference = inference
    
    def __len__(self):
        return len(self.lines)
    
    def __getitem__(self, idx):
        row = self.lines[idx][:-1].split(' ')
        
        fname = f'{self.path}/{row[0]}'
        img = cv2.cvtColor(cv2.imread(fname), cv2.COLOR_BGR2RGB)
        if self.inference:
            img = self.resize(image=img)#['image']
            return self.normalize(**img)['image'], img['image']
        
        num = int(row[1])
        boxes = []
        boxes = [[float(n) for n in row[i+1:i+5]] + [int(row[i])] for i in range(2, 2+5*num, 5)]
        for box in boxes:
            box[2] += box[0]
            box[3] += box[1]
        b = boxes[0]
        #if self.show:
        #    plt.scatter([box[0], box[2]], [box[1], box[3]], c='red')
        transformed = self.resize(image=img, bboxes=boxes)
        orig = transformed['image']
        if self.augment:
            transformed = self.transform(**transformed)
        out = self.normalize(**transformed)
        
        img = out['image']
        bboxes = torch.tensor(out['bboxes'])
        input_annots = np.ones((self.max_boxes, 5), dtype=np.float32) * (-1)
        input_annots[:bboxes.shape[0], :] = bboxes
        img = np.float32(np.transpose(img, [2,0,1]))
        if not self.show:
            return torch.tensor(img), torch.tensor(input_annots).int()
        return torch.tensor(img), orig
```

```python
train_ds = OktoberfestDataset(train_data, '../../../data/train', augment=True)
val_ds = OktoberfestDataset(val_data, '../../../data/train', augment=False)
test_ds = OktoberfestDataset(val_data, '../../../data/train', augment=False)
test_ds.show = True
#test_ds = OktoberfesetDataset(test, 'data/test', augment=False)
```

```python
from torch.utils.data import DataLoader
train_dl = DataLoader(train_ds, batch_size=16, shuffle=True)
val_dl = DataLoader(val_ds, batch_size=16)
```

```python
def iterate(dl, model, loss_fn, optimizer, update=False, training=False, scheduler=None):
    running_loss = 0
    cnt = 0
    acc_cnt = 0
    for x, y in dl:
        cnt += len(y)
        with torch.set_grad_enabled(training):
            out = model(x)
            #out['cls'] = torch.sigmoid(out['cls'])
            y = model.make_y(y)
            loss = model.losses(out, y)
        running_loss += loss
        if update:
            optimizer.zero_grad()
            loss.backward()
            optimizer.step()
    #print(out["cls"].max())
    return running_loss/cnt

def train(train, val, model, loss_fn, optimizer, scheduler=None, num_epochs=10, model_path='./model.pt'):
    best_loss = float('Inf')
    cnt = 0 # counter for how many epochs with out val improvement
    for i in range(num_epochs):
        if i == 2:
            for param in model.backbone.parameters():
                param.requires_grad=True
            for param in model.upsample.parameters():
                param.requires_grad=True
            optimizer = optim.Adam([{'params': model.backbone.parameters(), 'lr':1e-3/9},
                                    {'params': model.upsample.parameters(), 'lr':1e-3/6},
                                    {'params': model.head.parameters(), 'lr':1e-3}])
            scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=.5)
        model.train()
        train_loss = iterate(train, model, loss_fn, optimizer, update=True, training=True, scheduler=scheduler)
        #model.eval()
        val_loss = iterate(val, model, loss_fn, optimizer, update=False)
        if scheduler:
            scheduler.step(val_loss)
        if val_loss < best_loss:
            cnt = 0
            best_loss = val_loss
            torch.save(model.state_dict(), model_path)
        print(f'Epoch {i}: train loss: {train_loss:.4f}, val loss: {val_loss:.4f}')
        cnt += 1
        if cnt == 10:
            print('early stopping')
            break
            
    model.load_state_dict(torch.load(model_path))
```

```python
optimizer = optim.Adam(model.head.parameters(), lr=1e-4, weight_decay=1e-4)
#optimizer = optim.Adam([{'params': model.backbone.parameters(), 'lr':1e-3/9},
#                        {'params': model.upsample.parameters(), 'lr':1e-3/6},
#                        {'params': model.head.parameters(), 'lr':1e-3}])
scheduler = optim.lr_scheduler.ReduceLROnPlateau(optimizer, factor=.5)
train(train_dl, val_dl, model, None, optimizer, num_epochs=40, scheduler=scheduler)
```
